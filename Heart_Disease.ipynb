## Data Standarization

###### 1--> Defective Heart
###### 0--> Healthy Heart

#### Splitting the Features and Target

```python
x=heart_df.drop(columns='target',axis=1)
y=heart_df['target']
```

#### Splitting the data into training data and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
print(y.shape,y_train.shape,y_test.shape)
```

## Model Training

```python
model=LogisticRegression(max_iter=900)
```

```python
# Training the logistic regression model with training data
model.fit(x_train.values,y_train)
```

## Model Evaluation

```python
training_data_accuracy=model.predict(x_train.values)
training_accuracy=accuracy_score(training_data_accuracy,y_train)
print('Accuracy score on training data: ',training_accuracy)
```

```python
# Accuracy on testing data
testing_data_accuracy=model.predict(x_test.values)
test_accuracy=accuracy_score(testing_data_accuracy,y_test)
print('Accuracy score on testing data: ',test_accuracy)
```

### Building a Predictive system

```python
input_data=(55,0,1,132,342,0,1,166,0,1.2,2,0,2)
#changing input_data to numpy array
input_as_array=np.asarray(input_data)
#reshaping input data
input_reshaped=input_as_array.reshape(1,-1)
prediction = model.predict(input_reshaped)
if(prediction[0]==0):
    print("Patient does not have a heart disease")
else:
    print("Patient have a heart disease")
    
```
